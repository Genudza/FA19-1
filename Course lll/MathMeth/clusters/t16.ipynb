## Задача 3 (16:10) Транспортная задача (способ Фогеля)

Ограничения:
- a3->b2 запрещена
- a1->b4 ровно 70 ед



| $c_{i,j}$ | $b_1$ | $b_2$ | $b_3$ | $b_4$ |     |
|-----------|-------|-------|-------|-------|-----|
| $a_1$     | 5     | 14    | 6     | 21    | 120 |
| $a_2$     | 20    | 13    | 17    | 14    | 40  |
| $a_3$     | 8     | 21    | 6     | 7     | 185 |
|           | 95    | 110   | 80    | 75    |     |

$\sum a = 120 + 40 + 185 = 345$

$\sum b = 95 + 110 + 80 + 75 = 360$

$\sum a \neq \sum b$

Чтобы получить закрытую модель, введем доп груз равный 360—345 = 15.
