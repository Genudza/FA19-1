# Задача 3
<br>Вычилить дюрацию Маколея и модифицированную дюрацию облигации, заданной параметрами:</br>
- m = 4 года
- c = 8%
- F = 100 долл
- v = 1 
- $ i_{0} $ = 10%

Найти по дюрации Маколея приближенную оценку изменения цены облигации при снижении рыночной ставки на 5 б.п. (базисных пунктов)

Формулы:
- $ P_{0} - $ Оценка стоимости облигации
- *D* - Дюрация
- $ D^{*} $ - Дюрация модифицированная
- $ \triangle P $ - изменение цены

$ P_{0}=\sum_{t=1}^n\frac{C}{(1+r)^t}+\frac{F}{(1+r)^n} $

$ D=\frac{1}{P_{0}}(\sum_{t=1}^nt\frac{C}{(1+r)^t}+n\frac{F}{(1+r)^n}) $

$ D^{*}=\frac{D}{1+r} $

$ \frac{\triangle P}{P_{0}}= -D^{*}*\triangle r $

```python
n = 4 # m 
c = 0.08
F = 100
v = 1
r = 0.1 # i_0
bp = -0.0005 # базисный пункт
"""Считаем дюрацию модифицированную + ΔP"""

P0 = sum([c * F / (pow((1+r), t)) for t in range(1, n+1)]) + (F / pow(1+r, n))
D0 = (sum([c * F * t / (pow((1+r), t)) for t in range(1, n+1)]) + n*(F / pow(1+r, n))) * (1/P0)

print(f"P0 = {P0}\nD0 = {D0}")

DM = D0 / (1+r)
print(f"DM = {DM}")
delta_P = -DM * bp * P0
# ΔP изменение цены
print(f"ΔP = {delta_P}")
```

# Задача 4
Определить:
- a) точное
- b) приближенное процентное изменение цены облигации 

с параметрами:
- F = 100
- c = 7%
- m = 4%
- f = 1%
если начальная рыночная ставка 6% уменьшится на 1%

Формулы:
- $ P_{0} - $ Оценка стоимости облигации
- *D* - Дюрация
- $ D^{*} $ - Дюрация модифицированная
- $ \triangle P $ - изменение цены

$ P_{0}=\sum_{t=1}^n\frac{C}{(1+r)^t}+\frac{F}{(1+r)^n} $

$ D=\frac{1}{P_{0}}(\sum_{t=1}^nt\frac{C}{(1+r)^t}+n\frac{F}{(1+r)^n}) $

$ D^{*}=\frac{D}{1+r} $

$ \triangle P = -D^{*}*\triangle r $

```python
F = 100
c = 0.07
n = 4 # m
r = 0.06 #i_0 рыночная ставка
bp = -0.01 # уменьшится на 0.01
# f = 1, если f=2, то берем коэфф при f --> n*2 c/2
P0 = sum([c * F / (pow((1+r), t)) for t in range(1, n+1)]) + (F / pow(1+r, n))
D0 = (sum([c * F * t / (pow((1+r), t)) for t in range(1, n+1)]) + n * (F / pow(1+r, n))) * (1/P0)

print(f"P0 = {P0}\nD0 = {D0}")

# a) точное
DM = D0 / (1+r)
print(f"DM = {DM}")

# b) приближенное процентное изменение цены облигации с параметрами
delta_P = -DM * bp
print(f"ΔP = {delta_P}")
```

# Задача 6
<br>Рыночная ставка процента равна 10%. На рынке появилась облигация номиналом 50 000 рублей с 14% купоном с годовыми выплатами, </br>а ее погашение наступит ровно через 4 года.

На сколько процентов изменится цена (стоимость) облигации при уменьшении ставки процента на 1% (1 п.п)

Рассчитать точное изменение цены, а также оценку через дюрацию.

Формулы:
- $ P_{0} - $ Оценка стоимости облигации
- *D* - Дюрация
- $ D^{*} $ - Дюрация модифицированная
- $ \triangle P $ - изменение цены

$ P_{0}=\sum_{t=1}^n\frac{C}{(1+r)^t}+\frac{F}{(1+r)^n} $

$ D=\frac{1}{P_{0}}(\sum_{t=1}^nt\frac{C}{(1+r)^t}+n\frac{F}{(1+r)^n}) $

$ D^{*}=\frac{D}{1+r} $

$ \frac{\triangle P}{P_{0}}= -D^{*}*\triangle r $

```python
F = 50000
c = 0.14
n = 4
r = 0.1 #i_0 рыночная ставка
bp = -0.01 # уменьшится на 0.01

P0 = sum([c * F / (pow((1+r), t)) for t in range(1, n+1)]) + (F / pow(1+r, n))
# Дюрация
D0 = (sum([c * F * t / (pow((1+r), t)) for t in range(1, n+1)]) + n * (F / pow(1+r, n))) * (1/P0)

print(f"P0 = {P0}\nD0 = {D0}")

# точное изменение цены
DM = D0 / (1+r)
print(f"DM = {DM}")
```
